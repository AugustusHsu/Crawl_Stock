```python
import os
import pandas as pd
import numpy as np

#建立Model Information的Schema
if not os.path.isdir("../model"):
    os.makedirs("../model")
    
SchemaColumn = ['Name','Description','Sample']

SchemaName = ['FileName','ColumnList','WindowSize','NumOfPredictDay','Layer',
              'Loss','Optimizer','BatchSize','Epoch','ValidationSplit','ModelName']

SchemaDescription = ['檔案名稱','選擇的Column','以過去幾天的資料去做訓練、預測','要預測的天數','神經網路的層數',
                     'Loss Function的選擇','優化器的選擇','驗證集比例','訓練時Batch的大小',
                     '訓練幾次','儲存Model的名稱']

SchemaSample = ['2002.csv,3008.csv','["成交股數","成交金額","成交筆數","開盤價","最後揭示買價","最後揭示賣價","收盤價"]',
                '50','1,7,10,30','[len(ColumnList), 50, 100,NumOfPredictDay]','mse','adam','0.1','512',
                '10','0.h5,1.h5,2.h5']

ModelInformationSchema = pd.DataFrame(columns=SchemaColumn,index=range(len(SchemaName)))
for i in range(len(SchemaName)):
    ModelInformationSchema.iloc[i,0] = SchemaName[i]
    ModelInformationSchema.iloc[i,1] = SchemaDescription[i]
    ModelInformationSchema.iloc[i,2] = SchemaSample[i]
    #print(SchemaName[i],SchemaDescription[i],SchemaSample2[i])
    
ModelInformationSchema.to_csv('../model/ModelInformationSchema.csv')
#ModelInformationSchema
#len(SchemaName)
```

```python
Filename = "2002.csv"
ColumnList = ["成交股數","成交金額","成交筆數","開盤價","最後揭示買價","最後揭示賣價","收盤價"]
WindowSize = 50
NumOfPredictDay = 30

Layer = [len(ColumnList), 50, 100,NumOfPredictDay]
Loss = "mse"
Optimizer = "adam"

BatchSize = 512
Epoch = 5
ValidationSplit = 0.1

import os
DirList = list(os.listdir("../data"))[1:]

FilenameList = DirList
#ColumnList = []
WindowSizeList = [50,100]
NumOfPredictDayList = [3,7,10,30]
#Layer = []
#Loss = []
#Optimizer = []
#BatchSize = []
EpochList = [10]
#ValidationSplit = []
                
#建立Model Information

ModelInformationColumn = SchemaName

NumofIndex = len(FilenameList)*len(WindowSizeList)*len(NumOfPredictDayList)*len(EpochList)
ModelInformation = pd.DataFrame(columns=ModelInformationColumn,index=range(NumofIndex)) 

count = 0
for Filename in FilenameList:
    for WindowSize in WindowSizeList:
        for NumOfPredictDay in NumOfPredictDayList:
            for Epoch in EpochList:
                Layer = [len(ColumnList), 50, 100,NumOfPredictDay]
                ModelInformation.iloc[count,0] = Filename
                ModelInformation.iloc[count,1] = ColumnList
                ModelInformation.iloc[count,2] = WindowSize
                ModelInformation.iloc[count,3] = NumOfPredictDay
                ModelInformation.iloc[count,4] = Layer
                ModelInformation.iloc[count,5] = Loss
                ModelInformation.iloc[count,6] = Optimizer
                ModelInformation.iloc[count,7] = BatchSize
                ModelInformation.iloc[count,8] = Epoch
                ModelInformation.iloc[count,9] = ValidationSplit
                ModelInformation.iloc[count,10] = str(count) + ".h5"
                #print(count,end=' ')
                count += 1

ModelInformation.to_csv('../model/ModelInformation.csv')
ModelInformation.to_pickle('../model/ModelInformation.pickle')
ModelInformation
```
